This notebook computes *freeze-in* abundance for a specific dark matter candidate freezing in through fermions. See Sec. 5 of https://arxiv.org/abs/2309.02485

## Preamble and physical constants

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import kn, zeta
from scipy.interpolate import interp1d
from scipy.integrate import cumulative_trapezoid, quad


## Physical Constants ##
alpha= 1/137

m_e = 0.511 # MeV
muon_r = 206.768 # m_muon / m_e
tau_r = 3477.23 # m_tau / m_e
m_muon = muon_r*m_e # MeV
m_tau = tau_r*m_e # MeV

H_0 = 1e-39 # MeV
M_PL = 2.4e21 # MeV
h = 0.67
g_star = 10

c1 = np.sqrt(4*np.pi**3/45)
c2 = 2*np.pi**2/45

s0 = 2891.2 # entropy density today cm^-3
inv_cm = 1.23981e-4*1e-6 # MeV per cm^-1
s0 *= (inv_cm)**3

rho_crit = 3*H_0**2*M_PL**2
rho_DM = (.1/h**2)*rho_crit
```

# Compute cross section
where we use dimensionless variables

$$ x\equiv\frac{m_B}{T}; \ \ y\equiv\frac{m_f}{T}; \ \ z\equiv\frac{s}{T^2}; \ \ \& \ \ \tilde{\sigma}\equiv\sigma T^2$$

```python
def sigmatilde(z, y, x):
    betasq        = np.sqrt( (z-(y+x)**2.)*(z-(y-x)**2.) ) / z
    sigmatildeeg = ( betasq*( y**4.*(7.*z*x**2.-x**4.+24.*z**2.) + z*y**2.*(-37.*z*x**2.+2.*x**4.-16.*z**2.) - y**6.*(3.*x**2.+16.*z) + z**2.*(z*x**2.+7.*x**4.+4.*z**2.) + 4.*y**8. ) + 2.*z*x**2. * ( y**2.*(6.*z-10.*x**2.) - 2.*z*x**2. + 2.*x**4. + 9.*y**4. + z**2. ) * np.log ( (y**2.-x**2.+z*(1.+betasq)) / (y**2.-x**2.+z*(1.-betasq)) ) ) / ( z*x**2.*(z-y**2.)**3. )
    return sigmatildeeg

def sigma(s, m_f, m):
    return sigmatilde(s, m_f, m) # trick to return dimensionful version
```

### Set parameters

```python
m_B = 1e-3*m_e # MeV
m_f = 105*m_e # fermion mass
T_RH = 1.e2## reheat temperature MeV

x_i = m_B/T_RH # initial x
x_f = 4*m_B/m_f # final x
x_vals = np.logspace(np.log10(x_i), np.log10(x_f), 100)

s_min = (m_B + m_f)**2
s_max = 100*m_f**2
s_vals = np.logspace( np.log10(s_min), np.log10(s_max), 100)

T_vals = m_B / x_vals
z_vals = s_vals / T_vals**2 
y_vals = m_f / T_vals
```

### Plotting cross-sections

```python
cs = sigmatilde(z_vals, y_vals, x_vals)

plt.figure(dpi=150)
plt.title(r"Dimensionless Cross Section")
plt.plot(s_vals, cs, color="k")
plt.yscale("log")
plt.xscale("log")
plt.xlabel(r"$s$ [MeV$^2$]")
plt.ylabel(r"$\frac{\tilde{\sigma}}{\alpha g^2}$",rotation=0, labelpad=20,fontsize = 14)
# plt.legend()
plt.show()
```

# Comparing the quantity $\langle \sigma v \rangle n_{EQ}^2 \equiv R$
Using dimensionless quantities to build $R$, we have 

$$\langle \sigma v \rangle n_{EQ}^2 = R = \frac{m_B^4}{32 \pi^4 x^4}\int_{(x+y)^2}^{\infty} dz \tilde{\sigma}(z) K_1(\sqrt{z})\frac{(z-y^2)^2}{\sqrt{z}}$$


```python
def integrandegKR(z,x,y):
    betasq        = np.sqrt( (z-(y+x)**2.)*(z-(y-x)**2.) ) / z
    sigmatildeeg = ( betasq*( y**4.*(7.*z*x**2.-x**4.+24.*z**2.) + z*y**2.*(-37.*z*x**2.+2.*x**4.-16.*z**2.) \
    - y**6.*(3.*x**2.+16.*z) + z**2.*(z*x**2.+7.*x**4.+4.*z**2.) + 4.*y**8. ) \
    + 2.*z*x**2. * ( y**2.*(6.*z-10.*x**2.) - 2.*z*x**2. + 2.*x**4. + 9.*y**4. + z**2. ) \
    * np.log( (y**2.-x**2.+z*(1.+betasq)) / (y**2.-x**2.+z*(1.-betasq)) ) ) / ( z*x**2.*(z-y**2.)**3. )
    return sigmatildeeg * kn(1,np.sqrt(z)) / np.sqrt(z) * (z-y**2.)**2

def R(x, m, m_f):
    y = x*m_f/m
    prefactor = m**4/(32*np.pi**4*x**4)
    return prefactor*quad(integrandegKR, (x+y)**2 , np.inf, args=(x, y))[0] 
```

### Plot comparisons

```python
rate = []

for x in x_vals:
    rate.append(R(x, m_B, m_f))

plt.figure(dpi=150)
plt.title(r"$\frac{m_B}{m_f}=$ "+str(m_B/m_f))
plt.plot(x_vals, rate, color="k")
plt.yscale("log")
plt.xscale("log")
plt.xlabel(r"$x$")
plt.ylabel(r"$R$ [MeV$^4$]", labelpad=20,fontsize = 14)
# plt.legend()
# plt.savefig("sigmavn2-equal-masses-2.png", facecolor="white", bbox_inches="tight")
plt.show()
```

# Computing Dimensionless Production rate

$$ \frac{dY}{dx} = \frac{\langle \sigma v \rangle n^2}{x s(x) H(x)}$$

Define $s(x)$ and $H(x)$:

```python
# entropy density
def entropy(x, m):
    return c2*g_star*(m/x)**3

# Hubble
def H(x, m):
    return c1*np.sqrt(g_star)*(m/x)**2/M_PL
```

```python
def dYdx(x, m, m_f):
    return R(x, m, m_f) / (x * entropy(x, m) * H(x, m))
```

## Plotting

```python
dY = []

for x in x_vals:
    dY.append(dYdx(x, m_B, m_f))
```

```python
plt.figure(dpi=150)
plt.title(r"$\frac{m_B}{m_f}=$ "+str(m_B/m_f))
plt.plot(x_vals, dY, color="k")
plt.yscale("log")
plt.xscale("log")
plt.xlabel(r"$x$")
plt.ylabel(r"$\frac{dY}{dx}$",rotation=0, labelpad=20,fontsize = 14)
# plt.legend()
# plt.savefig("sigmavn2-equal-masses-2.png", facecolor="white", bbox_inches="tight")
plt.show()
```

# Finally, compute abundance

```python
def Abundance(x_array, m, m_f):
    dY = []
    for x in x_array:
        dY.append(dYdx(x, m, m_f))
    Y = cumulative_trapezoid(dY, x_array, initial=0 )
    return Y
```

```python
m_B = 1e-3*m_e # MeV
m_f = 1*m_e #fermion mass
T_RH = 1.e2## reheat temperature MeV

x_i = m_B/T_RH # initial x
x_f = 3*m_B/m_f # final x
x_vals = np.logspace(np.log10(x_i), np.log10(x_f), 100)


Y = Abundance(x_vals, m_B, m_f)

print("Y = ", Y[-1])
```

```python
plt.figure(dpi=150)
plt.title(r"$\frac{m_B}{m_f}=$ "+str(m_B/m_f))
plt.plot(x_vals, Y, color="k")
# plt.axhline(y=Y_asymptotic, label=r"$\frac{T_{RH}M_{P}}{m_B^2}$")
plt.yscale("log")
plt.xscale("log")
plt.xlabel(r"$x$")
plt.ylabel(r"$Y$",rotation=0, labelpad=20,fontsize = 14)
# plt.legend()
# plt.savefig("sigmavn2-equal-masses-2.png", facecolor="white", bbox_inches="tight")
plt.show()
```
